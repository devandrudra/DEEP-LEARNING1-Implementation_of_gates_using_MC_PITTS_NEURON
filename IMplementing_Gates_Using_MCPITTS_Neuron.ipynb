# OR GATE implementation

```python
# input : [x1, x2]
# output : [O]
# Thresold: 1
output_or = []
def OR_gate(input):
  for i in input:
    summation = sum(i)
    if summation >= 1:
      print(f"input: {i} -> output: 1")
      output_or.append(1)
    elif summation < 1:
      print(f"input: {i} -> output: 0")
      output_or.append(0)

input = [[0,0], [0,1], [1,0], [1,1]]
OR_gate(input)
```

# AND gate implementation

```python
# input : [x1, x2]
# output : [O]
# Thresold: 2

output_and = []
def AND_gate(input):
  for i in input:
    summation = sum(i)
    if summation >= 2:
      print(f"input: {i} -> output: 1")
      output_and.append(1)
    elif summation < 2:
      print(f"input: {i} -> output: 0")
      output_and.append(0)

input = [[0,0], [0,1], [1,0], [1,1]]
AND_gate(input)
```

# NAND GATE IMPLEMENTATION

```python
# input : [x1, x2]
# output : [O]
# Thresold: 1

output_nand = []
def NAND_gate(input):
  for i in input:
    global output_nand
    summation = sum(i)
    if summation >= 2:
      print(f"input: {i} -> output: 0")
      output_nand.append(0)
    elif summation < 2:
      print(f"input: {i} -> output: 1")
      output_nand.append(1)

input = [[0,0], [0,1], [1,0], [1,1]]
NAND_gate(input)
```

# NOR gate implementation

```python
# input : [x1, x2]
# output : [O]
# Thresold: 1

output_nor = []
def NOR_gate(input):
  for i in input:
    global output_nor
    summation = sum(i)
    if summation >= 2:
      print(f"input: {i} -> output: 1")
      output_nor.append(1)
    elif summation < 2:
      print(f"input: {i} -> output: 0")
      output_nor.append(0)

input = [[0,0], [0,1], [1,0], [1,1]]
NOR_gate(input)
```
